```python
df = pd.read_csv('tmdb_5000_movies.csv')
df .head(3)
```

```python
release_date_sample = df['release_date'].iloc[0]
release_date_sample[:4]
```

```python
def get_year(release_date):
    return int(release_date[:4])
```

```python
# lambda release_date: release_date[:4]
```

```python
get_year('2012-07-16')
```

```python
df[df['release_date'].isna()]
```

```python
df = df[~df['release_date'].isna()]
```

```python
df['year'] =  df['release_date'].apply(get_year)
```

```python
df.head(3)
```

```python
df.groupby('year').count()['id']
```

```python
plt.plot(df.groupby('year').count()['id'])
plt.xlabel('year')
plt.ylabel('count')
```

```python
df = df[df['revenue']!=0]
plt.plot(df.groupby('year').median()['revenue'], label='median')
plt.plot(df.groupby('year').mean()['revenue'], label='mean')
plt.plot(df.groupby('year')['revenue'].max(), label='max')
plt.legend()
```
